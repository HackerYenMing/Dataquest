```python
import pandas as pd
data = pd.read_csv('academy_awards.csv', encoding='ISO-8859-1')

data.head()
```

```python
# Use the value_counts method to explore 6 unnamed columns at the end.

data['Unnamed: 5'].value_counts()
```

```python
# filter our Dataframe to the same subset so it's more manageable.

data['Year'] = data['Year'].str[0:4]
data['Year'] = data['Year'].astype('int64')
# data['Year']

later_than_2000 = data[data['Year'] > 2000]
# later_than_2000['Year']

award_categories = ['Actor -- Leading Role', 'Actor -- Supporting Role',
                    'Actress -- Leading Role', 'Actress -- Supporting Role']
nominations = later_than_2000[later_than_2000['Category'].isin(award_categories)]

nominations.head()
```

```python
# Won? column -> replace all NO values with 0 and all YES values with 1.
# get rid of the 6 extra, unnamed columns

replace_dict = {'NO' : 0, 'YES' : 1}
nominations['Won'] = nominations['Won?'].map(replace_dict)

drop_col = ['Won?', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7',
            'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10']
final_nominations = nominations.drop(drop_col, axis=1)

final_nominations.head()
```

```python
additional_info_one = final_nominations['Additional Info'].str.rstrip("'}")
additional_info_two = additional_info_one.str.split(" {'")
# additional_info_two

final_nominations['Movie'] = additional_info_two.str[0]
final_nominations['Character'] = additional_info_two.str[1]
final_nominations.drop('Additional Info', axis=1)

final_nominations.head(2)
```
