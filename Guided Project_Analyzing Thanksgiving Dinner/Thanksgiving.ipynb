```python
import numpy as np
import pandas as pd

data = pd.read_csv('thanksgiving.csv', encoding="Latin-1")

#print(data.columns)
#print(data.head(5))
```

```python
count = pd.Series.value_counts(data['Do you celebrate Thanksgiving?'])
print(count)

data = data[data['Do you celebrate Thanksgiving?'] == 'Yes']
print(data.shape[0])
```

```python
apple_isnull = pd.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple'])
pumpkin_isnull = pd.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin'])
pecan_isnull = pd.isnull(data['Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan'])

ate_pies = apple_isnull & pumpkin_isnull & pecan_isnull
ate_count = pd.Series.value_counts(ate_pies)
print(ate_count)
```

```python
data['Age'].value_counts()
```

```python
def convert_age(item_age) :
    
    if pd.isnull(item_age) :
        return None
    
    first_item = item_age.split(' ')[0]
    first_item = first_item.replace('+', '')
    
    return int(first_item)

int_age = data['Age'].apply(convert_age)
data['int_age'] = int_age

print( data['int_age'].iloc[:11] )
print( data['int_age'].describe() )
```

```python
data['How much total combined money did all members of your HOUSEHOLD earn last year?'].value_counts()
```

```python
def convert_money(item_money) :
    
    if pd.isnull(item_money) :
        return None
    
    first_item = item_money.split(' ')[0]
    first_item = first_item.replace('$', '').replace(',', '')
    
    if first_item == 'Prefer' :
        return None
    return int(first_item)

int_income = data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(convert_money)
data['int_income'] = int_income

print( data['int_income'].iloc[:11] )
print( pd.Series.describe(data['int_income']) )
```
