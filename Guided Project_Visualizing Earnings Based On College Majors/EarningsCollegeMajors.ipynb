```python
# Use a grouped bar plot to compare the number of men with the number of women in each category of majors.
import numpy as np
import matplotlib.pyplot as plt

total_count = {}
cat_list = []
men_list = []
women_list = []

def count_gender_cat(row) :    
    cat = row['Major_category']
    men = int(row['Men'])
    women = int(row['Women'])
    
    if cat in total_count :
        total_count[cat][0] += men
        total_count[cat][1] += women
    else :
        total_count[cat] = [men, women]
    
def dictToList() :
    for cat, gender in total_count.items() :
        cat_list.append(cat)
        men_list.append(gender[0])
        women_list.append(gender[1])
                
category_table = recent_grads[['Major_category', 'Men', 'Women']]
category_table.apply(count_gender_cat, axis=1)
dictToList()
    
fig, ax = plt.subplots(figsize=(15, 10))

index = np.arange( len(cat_list) )
bar_width = 0.4
opacity = 0.5

rects1 = plt.barh(index, men_list, bar_width, 
                 alpha = opacity, color='b', label='Men')

rects2 = plt.barh(index + bar_width, women_list, bar_width, 
                 alpha =opacity, color='r', label='Women')

plt.ylabel('Men & Women Group')
plt.xlabel('Number of people')
plt.title('The number of men with the number of women in each category of majors')

plt.xticks(np.arange(15) * 50000)
plt.yticks(index + bar_width / 2, cat_list)
plt.legend()

plt.tight_layout()
plt.show()
```
